```python
import os
import cv2
import faiss
import time
import random
import numpy as np
from insightface.app import FaceAnalysis
from openpyxl import Workbook
from openpyxl.drawing.image import Image as ExcelImage

np.float = float

# =========================
# CONFIGURATION
# =========================

DATASET_PATH = "dataset"
REPORT_FILE = "attendance_report.xlsx"

SIM_THRESHOLD = 0.45
EMBEDDING_DIM = 512

LECTURE_DURATION = 60 * 60   # 1 hour
TOTAL_SAMPLES = 10
PRESENT_THRESHOLD = 7

CAMERA_SOURCES = [0]   # Add more: [0, 1, "rtsp://..."]

FRAME_WIDTH = 640

# =========================
# INITIALIZE MODEL
# =========================

def load_model():
    print("🔄 Loading InsightFace model...")
    app = FaceAnalysis(name="buffalo_l")
    app.prepare(ctx_id=0, det_thresh=0.35)
    return app

# =========================
# BUILD FAISS DATABASE
# =========================

def build_database(face_app):
    print("📚 Building face database...")

    index = faiss.IndexFlatIP(EMBEDDING_DIM)
    names = []
    embeddings = []
    student_photos = {}
    all_students = []

    for student in os.listdir(DATASET_PATH):
        folder = os.path.join(DATASET_PATH, student)

        if not os.path.isdir(folder):
            continue

        all_students.append(student)

        images = os.listdir(folder)
        if images:
            student_photos[student] = os.path.join(folder, images[0])

        for img_name in images:
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path)

            if img is None:
                continue

            faces = face_app.get(img)

            if not faces:
                continue

            emb = faces[0].embedding
            emb = emb / np.linalg.norm(emb)

            embeddings.append(emb)
            names.append(student)

    embeddings = np.array(embeddings).astype("float32")
    index.add(embeddings)

    print(f"✅ Loaded {len(names)} face embeddings")

    return index, names, all_students, student_photos

# =========================
# FACE RECOGNITION
# =========================

def recognize_face(index, names, embedding):
    embedding = embedding / np.linalg.norm(embedding)
    embedding = embedding.astype("float32").reshape(1, -1)

    scores, ids = index.search(embedding, 1)

    if scores[0][0] > SIM_THRESHOLD:
        return names[ids[0][0]]

    return "Unknown"

# =========================
# PROCESS FRAME
# =========================

def process_frame(frame, face_app, index, names):
    detected = []

    h, w = frame.shape[:2]
    scale = FRAME_WIDTH / max(h, w)

    small = cv2.resize(frame, (int(w * scale), int(h * scale)))
    faces = face_app.get(small)

    for face in faces:
        emb = face.embedding
        name = recognize_face(index, names, emb)

        if name != "Unknown":
            detected.append(name)

    return detected

# =========================
# RANDOM SAMPLING TIMES
# =========================

def generate_random_times(start, duration, n):
    return sorted([start + random.uniform(0, duration) for _ in range(n)])

# =========================
# MAIN ATTENDANCE LOGIC
# =========================

def run_attendance(face_app, index, names, all_students):

    print("🎥 Initializing cameras...")
    caps = [cv2.VideoCapture(src) for src in CAMERA_SOURCES]

    attendance_counts = {student: 0 for student in all_students}

    start_time = time.time()
    check_times = generate_random_times(start_time, LECTURE_DURATION, TOTAL_SAMPLES)

    print("🚀 Starting smart attendance system...")

    for i, t in enumerate(check_times):

        while time.time() < t:
            time.sleep(0.5)

        print(f"📸 Sample {i+1}/{TOTAL_SAMPLES}")

        frame_results = []

        for cap in caps:
            ret, frame = cap.read()
            if not ret:
                continue

            names_detected = process_frame(frame, face_app, index, names)
            frame_results.extend(names_detected)

        # MULTI-CAMERA FUSION
        unique_names = set(frame_results)

        for name in unique_names:
            attendance_counts[name] += 1

    for cap in caps:
        cap.release()

    return attendance_counts

# =========================
# GENERATE REPORT
# =========================

def generate_report(attendance_counts, all_students, student_photos):

    wb = Workbook()
    ws = wb.active
    ws.title = "Attendance"

    ws.append(["Student", "Detected Count", "Status", "Photo"])

    row = 2

    for student in all_students:

        count = attendance_counts.get(student, 0)

        status = "Present" if count >= PRESENT_THRESHOLD else "Absent"

        ws.cell(row=row, column=1, value=student)
        ws.cell(row=row, column=2, value=count)
        ws.cell(row=row, column=3, value=status)

        photo_path = student_photos.get(student)

        if photo_path and os.path.exists(photo_path):
            img = ExcelImage(photo_path)
            img.width = 80
            img.height = 80
            ws.add_image(img, f"D{row}")

        row += 1

    wb.save(REPORT_FILE)
    print(f"📊 Report saved: {REPORT_FILE}")

# =========================
# MAIN ENTRY POINT
# =========================

def main():

    face_app = load_model()

    index, names, all_students, student_photos = build_database(face_app)

    attendance_counts = run_attendance(face_app, index, names, all_students)

    generate_report(attendance_counts, all_students, student_photos)

    print("✅ Attendance process completed")

# =========================

if __name__ == "__main__":
    main()
```
